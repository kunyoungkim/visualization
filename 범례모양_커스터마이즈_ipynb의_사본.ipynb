```python
import matplotlib.pyplot as plt
import numpy as np

# Create example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

# Plot the data
plt.figure(figsize=(9, 6))
plt.plot(x, y1, label='sin(x)', linestyle='-', linewidth=2, color='blue')
plt.plot(x, y2, label='cos(x)', linestyle='--', linewidth=2, color='green')
plt.plot(x, y3, label='sin(x) + cos(x)', linestyle='-.', linewidth=2, color='red')

# Add legend
plt.legend()

# Add title and labels
plt.title('Example Plot for plt.legend()')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```

```python
# Create the plot
plt.figure(figsize=(10, 6))
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot the data
plt.plot(x, y1, label='sin(x)', linestyle='-', linewidth=2, color='blue')
plt.plot(x, y2, label='cos(x)', linestyle='--', linewidth=2, color='green')

# Add some text to the plot
plt.text(5, 0, "Important Point", backgroundcolor='yellow')

# Add legend with 'best' location
plt.legend(loc='best')

# Add title and labels
plt.title('Example with loc="best" and Text')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```

```python
# Create a 2x2 layout
fig = plt.figure(figsize=(9, 7))

# Create subplot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'b-', label='line1')
plt.title("plt.legend(loc='upper right')")
plt.legend(loc='upper right')

# Create subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, y1, 'g--', label='line2')
plt.title("plt.legend(loc='lower left')")
plt.legend(loc='lower left')

# Create subplot 3
plt.subplot(2, 2, 3)
plt.plot(x, y1, 'r-.', label='line3')
plt.title("plt.legend(loc='center right')")
plt.legend(loc='center right')

# Create subplot 4
plt.subplot(2, 2, 4)
plt.plot(x, y1, 'm:', label='line4')
plt.title("plt.legend(loc='lower center')")
plt.legend(loc='lower center')

# Add spacing between subplots for better visibility
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Create example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the plot
plt.figure(figsize=(9, 5))
plt.plot(x, y1, label='sin(x)', linestyle='-', linewidth=2, color='blue')
plt.plot(x, y2, label='cos(x)', linestyle='--', linewidth=2, color='green')

# Add a custom legend with various options
plt.legend(
    title='My Legend Title',
    title_fontsize=16,
    loc='upper left',
    bbox_to_anchor=(0,0),
    borderpad=2,
    labelspacing=1,
    handlelength=3,
    handletextpad=2,
    borderaxespad=2,
    columnspacing=2,
    ncol=2,
    fontsize='small',
    frameon=True,
    shadow=True,
    prop={'weight': 'normal'}
)

# Add title and labels
plt.title('Example Plot with Customized Legend')
plt.xlabel('x')
plt.ylabel('y')

# Adjust layout to make room for the legend
plt.tight_layout(rect=[0, 0, 0.75, 1])

# Show the plot
plt.show()
```
