```python
# 데이터 생성
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [200, 220, 250, 275, 300, 320]
profit_rate = [0.1, 0.15, 0.12, 0.2, 0.22, 0.18]

# 이중축 그래프 그리기
fig, ax1 = plt.subplots(figsize=(9, 6))

ax2 = ax1.twinx()
sns.barplot(x=months, y=revenue, ax=ax1, palette='coolwarm', alpha=0.6)
sns.lineplot(x=months, y=profit_rate, ax=ax2, color='b', marker='o')

# 축 레이블과 타이틀 추가
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue', color='g')
ax2.set_ylabel('Profit Rate', color='b')
plt.title('Monthly Revenue and Profit Rate')

# 그래프 보이기
plt.show()
```

```python
# 3D 막대 차트 그리기
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

x = np.arange(len(categories))  # x 위치 설정
y = np.zeros(len(categories))  # y 위치는 모두 0으로 설정
z = np.zeros(len(categories))  # z 위치 (바닥)은 모두 0으로 설정
dx = np.ones(len(categories)) * 0.5  # 바의 너비
dy = np.ones(len(categories)) * 0.5  # 바의 깊이
dz = values  # 바의 높이

ax.bar3d(x, y, z, dx, dy, dz, shade=True)

# 축 레이블과 타이틀 추가
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Bar Chart Example')

# 그래프 보이기
plt.show()
```

```python
# NumPy 배열로 변환
x = np.array(x)
y = np.array(y)
z = np.array(z)
dz = np.array(dz)

# 3D 막대 차트 그리기 with 동일한 축 간격
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# 3D 막대 차트 그리기
ax.bar3d(x, y, z, dx, dy, dz, shade=True)

# 축 레이블과 타이틀 추가
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Bar Chart with Equal Axis Spacing')

# x, y, z 축 간격을 동일하게 설정
max_range = np.array([x.max()-x.min(), y.max()-y.min(), z.max()+dz.max()-z.min()]).max() / 2.0

mid_x = (x.max()+x.min()) * 0.5
mid_y = (y.max()+y.min()) * 0.5
mid_z = (z.max()+dz.max()+z.min()) * 0.5

ax.set_xlim(mid_x - max_range, mid_x + max_range)
ax.set_ylim(mid_y - max_range, mid_y + max_range)
ax.set_zlim(mid_z - max_range, mid_z + max_range)

# plt.ylim(0,1)

# 그래프 보이기
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

# 임의의 데이터 생성
np.random.seed(0)
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10),
    'D': np.random.rand(10),
    'E': np.random.rand(10)
}

df = pd.DataFrame(data)

# 상관계수 행렬 계산
correlation_matrix = df.corr()

# 상관계수 표 출력
print("Correlation Matrix:")
print(correlation_matrix)

# 히트맵 그리기 with only xlabel and no x-axis line
plt.figure(figsize=(9, 3))
ax = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
# plt.title("Correlation Heatmap")

# x축 레이블을 그래프 위에 표시
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')

# x축과 y축의 라인을 제거
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# y축 레이블 회전 설정
plt.yticks(rotation=0)

plt.show()
```
